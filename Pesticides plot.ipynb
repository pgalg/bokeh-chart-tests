```python
import pandas
df=pandas.read_excel("pesticides.xlsx", sheetname=0)
list(df)
```

```python
import os
os.listdir()
```

```python
from bokeh.plotting import figure, output_file, show


plotdf=pandas.DataFrame()
plotdf['Ecotox']=df['TETP(E)']
plotdf['Humtox']=df['HTP(E) End HH']
plotdf['Name']=df['Substance name (ReCiPe)']
plotdf['Compartment']=df["Compartments"]+", "+df["Subcompartment"]

plotdf.iloc[0:4,0:4]
```

```python
from bokeh.plotting import ColumnDataSource

dataplot=ColumnDataSource(plotdf)
```

```python
from bokeh.models import HoverTool, ResetTool, WheelZoomTool, PanTool, BoxZoomTool
hover=HoverTool(
     tooltips=[
            ("Pollutant Name", "@Name"),
            ("Compartment", "@Compartment"),
            ("Ecotoxicity", "$x kg 1,4db eq/kg"),
            ("Human toxicity", "$y DALY/kg")
        ]
    )
```

```python
b=figure(title="Toxic emissions",  tools=[hover, ResetTool(), WheelZoomTool(),PanTool(), BoxZoomTool()])

b.circle('Ecotox','Humtox', size=5, color="grey", source=dataplot)

b.xaxis.axis_label="Ecotoxicity (kg 1,4db eq/kg)"
b.yaxis.axis_label="Human toxicity (DALY/kg)"
```

```python
from bokeh.io import output_notebook
output_notebook()
output_file("pesticides plot.html")
# initializes the output file

show(b)
# add b to file
```
